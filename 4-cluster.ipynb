```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
import pandas as pd
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
df
```

```python
df.shape
```

```python
df["target"] = iris["target"]
df
```

```python
from sklearn.model_selection import train_test_split
# train_test_split()=> æœƒç”¢ç”Ÿä¸€å€‹tuple,å¦‚(ç‰¹å¾µ90%, ç‰¹å¾µ10%, ç›®æ¨™90%, ç›®æ¨™10%)
x_train, x_test, y_train, y_test = train_test_split(iris["data"], \
                                                    iris["target"], test_size=0.1)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_test
```

```python
# KMeansæ˜¯å€‹ç‰©ä»¶,ç‰©ä»¶é€šå¸¸ç¬¬ä¸€å€‹å­—æœƒå¤§å¯«
# fit è¨“ç·´ä¸¦åšå‡ºå…¬å¼, æ³¨æ„:éžç›£ç£å¼å­¸ç¿’ è¨“ç·´æ™‚ä¸¦æ²’çµ¦ç­”æ¡ˆ
from sklearn.cluster import KMeans
clu = KMeans(n_clusters=3)
clu.fit(x_train)
```

```python
# å¹«æˆ‘åŠ æ¨™ç±¤(å“ªä¸€ç¾¤) Labels of each point
clu.labels_
```

```python
pre = clu.predict(x_test)
print("é æ¸¬çµæžœ:", pre)
print("æ­£ç¢ºç­”æ¡ˆ:", y_test)
```

```python
#é æ¸¬ç‚º0 ä¸ä¸€å®šæ˜¯æ­£ç¢ºç­”æ¡ˆè£¡çš„0åˆ†é¡ž, å¯å¤šåŸ·è¡Œå¹¾æ¬¡
from sklearn.metrics import accuracy_score
score = accuracy_score(pre, y_test)
print("æ­£ç¢ºçŽ‡:", score)
```

## ====éžç›£ç£å¼å­¸ç¿’, ç•¶ä¸çŸ¥é“kæœ‰å¹¾é¡žæ™‚===

```python
# axis=1åˆªé™¤è¡Œ(æ¬„), 0åˆªé™¤åˆ—
x = df.drop(["target"], axis=1)
x
```

```python
# çœ‹è¬›ç¾© Silhouette Score
from sklearn.metrics import silhouette_score
plotx = []
ploty = []
for k in range(2,6):
    clu = KMeans(n_clusters=k)
    clu.fit(x)
    s = silhouette_score(x, clu.labels_)
    plotx.append(k)
    ploty.append(s)
print("åˆ†å¹¾ç¾¤:", plotx)
print("scoreç‚º:", ploty)
```

```python
#æ­é…çœ‹ demo2 classification treeåœ–,æˆ‘å€‘å¯ä»¥åˆç†çš„æŽ¨æ¸¬å…¶å¯¦é€™ä¸‰é¡žä¸­æœ‰å…©é¡žæ˜¯å¾ˆåƒçš„ï¼
```

```python
import matplotlib.pyplot as plt
plt.plot(plotx, ploty)
plt.title("Silhouette Score")
plt.xlabel("K")
plt.ylabel("Score")
plt.show()
```
