```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
import pandas as pd
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
df
```

```python
df.shape
```

```python
df["target"] = iris["target"]
df
```

```python
from sklearn.model_selection import train_test_split
# train_test_split()=> 會產生一個tuple,如(特徵90%, 特徵10%, 目標90%, 目標10%)
x_train, x_test, y_train, y_test = train_test_split(iris["data"], \
                                                    iris["target"], test_size=0.1)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_test
```

```python
# KMeans是個物件,物件通常第一個字會大寫
# fit 訓練並做出公式, 注意:非監督式學習 訓練時並沒給答案
from sklearn.cluster import KMeans
clu = KMeans(n_clusters=3)
clu.fit(x_train)
```

```python
# 幫我加標籤(哪一群) Labels of each point
clu.labels_
```

```python
pre = clu.predict(x_test)
print("預測結果:", pre)
print("正確答案:", y_test)
```

```python
#預測為0 不一定是正確答案裡的0分類, 可多執行幾次
from sklearn.metrics import accuracy_score
score = accuracy_score(pre, y_test)
print("正確率:", score)
```

## ====非監督式學習, 當不知道k有幾類時===

```python
# axis=1刪除行(欄), 0刪除列
x = df.drop(["target"], axis=1)
x
```

```python
# 看講義 Silhouette Score
from sklearn.metrics import silhouette_score
plotx = []
ploty = []
for k in range(2,6):
    clu = KMeans(n_clusters=k)
    clu.fit(x)
    s = silhouette_score(x, clu.labels_)
    plotx.append(k)
    ploty.append(s)
print("分幾群:", plotx)
print("score為:", ploty)
```

```python
#搭配看 demo2 classification tree圖,我們可以合理的推測其實這三類中有兩類是很像的！
```

```python
import matplotlib.pyplot as plt
plt.plot(plotx, ploty)
plt.title("Silhouette Score")
plt.xlabel("K")
plt.ylabel("Score")
plt.show()
```
