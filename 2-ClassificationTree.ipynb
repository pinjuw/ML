```python
from sklearn.model_selection import train_test_split
# train_test_split()=> æœƒç”¢ç”Ÿä¸€å€‹tuple,å¦‚(ç‰¹å¾µ90%, ç‰¹å¾µ10%, ç›®æ¨™90%, ç›®æ¨™10%)
x_train, x_test, y_train, y_test = train_test_split(iris["data"], \
                                                    iris["target"], test_size=0.1)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_test
```

```python
# DecisionTreeClassifieræ˜¯å€‹ç‰©ä»¶,ç‰©ä»¶é€šå¸¸ç¬¬ä¸€å€‹å­—æœƒå¤§å¯«
# fit è¨“ç·´ä¸¦åšå‡ºå…¬å¼
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
pre = clf.predict(x_test)
print("é æ¸¬çµæžœ:", pre)
print("æ­£ç¢ºç­”æ¡ˆ:", y_test)
```

```python
print("é æ¸¬çµæžœ:", list(pre))
print("æ­£ç¢ºç­”æ¡ˆ:", list(y_test))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(pre, y_test)
print("æ­£ç¢ºçŽ‡:", score)
```

```python
# DecisionTreeClassifieræ˜¯å€‹ç‰©ä»¶,ç‰©ä»¶é€šå¸¸ç¬¬ä¸€å€‹å­—æœƒå¤§å¯«
# éŽæ“¬åˆä¸è¦‹å¾—å¥½, å¯èª¿ max_depth=
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(x_train, y_train)
```
