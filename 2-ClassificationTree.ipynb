```python
from sklearn.model_selection import train_test_split
# train_test_split()=> 會產生一個tuple,如(特徵90%, 特徵10%, 目標90%, 目標10%)
x_train, x_test, y_train, y_test = train_test_split(iris["data"], \
                                                    iris["target"], test_size=0.1)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_test
```

```python
# DecisionTreeClassifier是個物件,物件通常第一個字會大寫
# fit 訓練並做出公式
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
pre = clf.predict(x_test)
print("預測結果:", pre)
print("正確答案:", y_test)
```

```python
print("預測結果:", list(pre))
print("正確答案:", list(y_test))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(pre, y_test)
print("正確率:", score)
```

```python
# 請先在pc與pycharm安裝好grahpviz
from sklearn.tree import export_graphviz
# 產生敘述
g = export_graphviz(clf, out_file=None, feature_names=iris["feature_names"], \
                class_names=iris["target_names"], filled=True)
```

```python
import graphviz
#把敘述產生圖
graphviz.Source(g)
```

```python
# DecisionTreeClassifier是個物件,物件通常第一個字會大寫
# 過擬合不見得好, 可調 max_depth=
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2)
clf.fit(x_train, y_train)
```

```python
# 請先在pc與pycharm安裝好grahpviz
from sklearn.tree import export_graphviz
# 產生敘述
g = export_graphviz(clf, out_file=None, feature_names=iris["feature_names"], \
                class_names=iris["target_names"], filled=True)
```

```python
import graphviz
#把敘述產生圖
graph = graphviz.Source(g)
graph
```
